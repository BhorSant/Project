#### Chart - 3 Histplot

```python
numeric_features = df.describe().columns
```

```python
from matplotlib.lines import lineStyles


for col in numeric_features[1:]:
  fig = plt.figure(figsize=(9,6))
  ax = fig.gca()
  feature = df[col]
  feature.hist(bins=50,ax=ax)
  ax.axvline(feature.mean(), color='magenta', linestyle='dashed', linewidth=2)
  ax.axvline(feature.median(), color='cyan', linestyle='dashed', linewidth=2)  

  ax.set_xlabel(col + " counts")
  ax.set_ylabel("Frequency")
  ax.set_title(col + " distribution")
  
  plt.show()
```

##### Que 1. Why did you pick the specific chart?

1. **Distribution Visualization:** Histograms are effective for displaying the distribution of numeric data, allowing us to observe patterns, peaks, and potential outliers.

2. **Central Tendency Comparison:** The vertical dashed lines at the mean (magenta) and median (cyan) provide a clear comparison of central tendency measures within each numeric feature, helping to assess the skewness or symmetry of the distribution.

3. **Bin Configuration:** The use of 50 bins in the histogram provides a detailed view of the data distribution, helping to identify specific ranges where the data is more concentrated.

4. **Comparison Across Features:** By iterating through multiple numeric features in a loop, this approach enables a quick comparison of the distributions of different variables in the dataset.

5. **Insight into Data Characteristics:** Histograms facilitate a visual understanding of the data characteristics, such as whether the data is normally distributed, skewed, or exhibits other patterns.
